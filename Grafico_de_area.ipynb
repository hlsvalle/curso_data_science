```
#Importando as bibliotecas necessarias
import pandas as pd
import matplotlib.pyplot as plt
```

```
#Importando nosso dataset
data = "https://raw.githubusercontent.com/carlosfab/curso_data_science_na_pratica/master/modulo_03/dengue-dataset.csv"
df = pd.read_csv(data, index_col='data', parse_dates=True) #Transformando a coluna data em index, importando nosso dataset para um dataframe
```

```
df.head() #Verificando as 5 primeiras linhas do nosso dataframe
```

```
fig, ax = plt.subplots(figsize=(10,6))

df.loc['1999-01-01' : '2001-01-01', ['temperatura-mininima',
                                     'temperatura-media',
                                     'temperatura-maxima']].plot.area(ax=ax)
plt.tight_layout();                                      
```

```
#Casos Confirmados X Temperatura-Media

df_norm = (df - df.mean()) / df.std() #Normalizando o dataframe

fig, ax = plt.subplots(figsize=(10,6))


#df_norm.loc['1999-01-01' : '2001-01-01','temperatura-media'].plot(kind='area', ax=ax)
df_norm.loc['1999-01-01' : '2001-01-01','casos-confirmados'].plot(kind='area', ax=ax)
ax.set_title("Casos confirmados X Temperatura Mdia")

plt.tight_layout();
```
