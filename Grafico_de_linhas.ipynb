### Gráfico de Linhas

```
#Importando as Bibliotecas Necessarias
import pandas as pd
import matplotlib.pyplot as plt
```

```
#Obtenção dos Dados
data = "https://raw.githubusercontent.com/carlosfab/curso_data_science_na_pratica/master/modulo_03/dengue-dataset.csv"
df = pd.read_csv(data, index_col='data', parse_dates=True)

#Verificando as 5 primeiras Linhas do DataFrame
df.head()
```

```
#Plotar os casos confirmados
fig, ax = plt.subplots()
df['casos-confirmados'].plot(kind='line', ax=ax)
ax.set_title("Dengue em Campinas")
ax.set_xlabel("Data")
ax.set_ylabel("Casos Confirmados")

plt.tight_layout();
```

```
#Selecionando Intervalo do DataFrame
fig, ax = plt.subplots()
df.loc['2007-01-01':'2014-12-01','temperatura-media'].plot(kind='line', ax=ax)
ax.set_title("Temperatura Média entre 2007 até 2014")
ax.set_xlabel("Data")
ax.set_ylabel("Temperatura Média")

plt.tight_layout()
```

```
#Plotar mais de um Gráfico Separadamente
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
df.loc['2007-01-01':'2014-12-01','temperatura-media'].plot(kind='line', ax=ax[0])
ax[0].set_title("Temperatura Média entre 2007 até 2014")
ax[0].set_xlabel("Data")
ax[0].set_ylabel("Temperatura Média")

df.loc['2007-01-01':'2014-12-01','casos-confirmados'].plot(kind='line', ax=ax[1])
ax[1].set_title("Casos Confirmados em Campinas")
ax[1].set_xlabel("Data")
ax[1].set_ylabel("Casos Confirmados")

plt.tight_layout();
```

```
#Plotar os graficos sobrepostos
df_norm = (df - df.mean()) / df.std() #Normalizando o Dataframe

#Plotar os graficos
fig, ax = plt.subplots(figsize=(8,4))

df_norm.loc['2007-01-01':'2014-12-01','temperatura-media'].plot(kind='line', ax=ax, color='green')
df_norm.loc['2007-01-01':'2014-12-01','casos-confirmados'].plot(kind='line', ax=ax)

ax.set_title("Casos confirmados X Temperatura Média")
ax.set_xlabel("Data")
ax.legend(loc='best')


plt.tight_layout()
```
