```
#Importando as bibliotecas necessarias
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = "https://raw.githubusercontent.com/carlosfab/data_science/master/datasets/housing.csv" #Importando o dataset

df = pd.read_csv(data)

df.head() #Verificando as 5 primeiras linhas do nosso dataframe
```

```
fig, ax = plt.subplots(figsize=(10,6))

df.plot.scatter(x='median_income', y='median_house_value',ax=ax, s=1)

plt.tight_layout()
```

# Regressão Linear em um Scatter Plot

```
#Importando o Seaborn
import seaborn as sns

fig, ax = plt.subplots()

ax.set_xlim(right=8)

sns.regplot('median_income', 'median_house_value', data=df,scatter_kws={"s": 0.2},line_kws={'color':'green'},ax=ax)
            
plt.tight_layout()
```
