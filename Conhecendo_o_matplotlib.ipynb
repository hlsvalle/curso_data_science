```
#importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```
#Abordagem Stateful
x = np.arange(10)
y = np.random.normal(size=10)

plt.plot(x,y);
plt.title("Titulo")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```
#Abordagem Stateless
x = np.arange(10)
y = np.random.normal(size=10)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title("Titulo")
ax.set_xlabel("x")
ax.set_ylabel("y")

plt.tight_layout()
```

```
#Abordagem Stateless com 2 Axes em 1 Figure

x1 = np.arange(10)
y1 = np.random.normal(size=10)

x2 = np.arange(10)
y2 = 1 / np.exp(np.random.normal(size=10))

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,3))
ax[0].plot(x1,y1)
ax[0].set_title("Titulo 1")
ax[0].set_xlabel("x1")
ax[0].set_ylabel("y1")

ax[1].plot(x2,y2)
ax[1].set_title("Titulo 2")
ax[1].set_xlabel("x2")
ax[1].set_ylabel("y2")

plt.tight_layout()
```
