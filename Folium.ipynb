```
#Importando as bibliotecas necessarias
import folium
import pandas as pd
```

```
mapa = folium.Map(location=[-20.4970485,-54.6204637], zoom_start=11, width=800, height=400, position='relative')
mapa
```

```
data = "https://raw.githubusercontent.com/carlosfab/curso_data_science_na_pratica/master/modulo_03/estacoesbike.csv"
df = pd.read_csv(data, delimiter=';')
```

```
latitude = df.latitude.values
longitute = df.longitude.values
nome = df.nome.values
```

```
mapa = folium.Map(location=[-8.0648734,-34.8703724], zoom_start=11)
```

```
for lat, lon, nome in zip(latitude, longitute, nome):
  folium.Marker(location=[lat, lon], popup=nome).add_to(mapa)

mapa
```
